```python
#QUESTION3

numbers = [45, 22, 89, 10, 66]

max = 0
min = 0

for i in range(len(numbers)):
  if max < numbers[i]:
    max = numbers[i]

  if min > numbers[i]:
    min = numbers[i]

print("Max:",max)
print("Min:",min)
```

```python
#QUESTION4

prices = [450, 1200, 899, 1500, 300]

count = 0
for i in range(len(prices)):
  if prices[i] > 1000:
    count +=1

print("Products above 1000:", count)
```

```python
#QUESTION5

attendance = ["P", "P", "A", "P", "P"]

absent = 0

for i in range(len(attendance)):
  if attendance[i] == "A":
    absent +=1

present = len(attendance) - absent
print("Attendance percentage: ", present/len(attendance)*100)
```

```python
#QUESTION7

text = "pythonP"

ch_count = {}

for c in text:
  if c in ch_count:
    ch_count[c] += 1
  else:
    ch_count[c] = 1

print(ch_count)
```
