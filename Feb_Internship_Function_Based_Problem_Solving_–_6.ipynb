```python
def smart_parking(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            if current > 0:
                current -= 1

        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)
    print("Peak Parking Usage:", peak)

    if current > capacity:
        print("Parking Status: Exceeded Capacity!")
    else:
        print("Parking Status: Available")


# Taking input from user
capacity = int(input("Enter Parking Capacity: "))
logs = input("Enter Vehicle Logs (space separated IN/OUT): ").split()

smart_parking(capacity, logs)
```

```python
def delivery_time(distance, traffic, weather):

    # assume normal speed is 25 km per hour
    time = (distance / 25) * 60   # convert to minutes

    # add delay based on traffic
    if traffic == "Low":
        time = time + 5
    elif traffic == "Medium":
        time = time + 10
    elif traffic == "High":
        time = time + 20

    # add delay based on weather
    if weather == "Cloudy":
        time = time + 5
    elif weather == "Rainy":
        time = time + 16
    elif weather == "Storm":
        time = time + 25

    return int(time)


# taking input from user
distance = float(input("Distance (km): "))
traffic = input("Traffic Level: ")
weather = input("Weather: ")

result = delivery_time(distance, traffic, weather)

print("Estimated Delivery Time:", result, "minutes")
```

```python
def seat_analyzer(total_seats, booked_list):

    booked_count = len(booked_list)

    # calculate percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    # decide show status
    if occupancy == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open Additional Show")

    elif occupancy >= 75:
        print("Show Status: Almost Full")

    elif occupancy >= 50:
        print("Show Status: Filling Fast")

    else:
        print("Show Status: Seats Available")


# taking input from user
total = int(input("Total Seats: "))

# user enters seats like: 1 1 1 1 1 ...
booked = list(map(int, input("Booked Seats (space separated): ").split()))

seat_analyzer(total, booked)
```

```python
def server_load(cpu_list):

    total = 0

    # calculate total using loop
    for value in cpu_list:
        total = total + value

    # calculate average
    average = total / len(cpu_list)

    print("Average CPU Load:", int(average), "%")

    # classify load
    if average < 50:
        print("Server Status: Normal")

    elif average >= 50 and average <= 80:
        print("Server Status: Warning")

    else:
        print("Server Status: Critical")


# taking input from user
readings = list(map(int, input("CPU Readings (%): ").split()))

server_load(readings)
```

```python
def resource_monitor(resources):

    overused = []

    # set limit (example: 8 hours per day)
    limit = 8

    for item in resources:
        if resources[item] > limit:
            overused.append(item)

    if len(overused) > 0:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# sample input (can also take from user)
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

```python
def event_registration(capacity, total_registrations):

    confirmed = 0
    waitlist = 0

    # simulate each registration
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed = confirmed + 1
        else:
            waitlist = waitlist + 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)

    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# taking input from user
capacity = int(input("Event Capacity: "))
registrations = int(input("Registrations: "))

event_registration(capacity, registrations)
```
