```python
class smartl:
  def __init__(self, name):
    self.name = name
    self.status = "OFF"

  def turn_on(self):
    self.status = "ON"

  def turn_off(self):
    self.status = "OFF"

  def print_status(self):
    print(f"{self.name} is {self.status}")

l_name = input("Enter the name of the light: ")
act = input("Enter the action (on/off): ")

l = smartl(l_name)

if act.upper() == "ON":
  l.turn_on()
elif act.upper() == "OFF":
  l.turn

l.print_status()
```

```python
class employee:

  def __init__(self, name, id, dept):
    self.name = name
    self.id = id
    self.dept = dept

  def print_emp(self):
    print("Employee ID Card")
    print(f"Name: {self.name}")
    print(f"ID: {self.id}")
    print(f"Department: {self.dept}")



ename = input("Employee name:")
eid = input("Employee id:")
edept = input("Department")

a = employee(ename, eid, edept)
a.print_emp()
```

```python
class contact:
  def __init__(self, name, no):
    self.name = name
    self.no = no

  def print_contact(self):
    print("Contact saved")
    print(f"Name: {self.name}")
    print(f"Number: {self.no}")


name = input("Contact name:")
no = input("Phone number:")

a = contact(name, no)
a.print_contact()
```

```python
class Movie:
    def __init__(self, name, r):
        self.name = name
        self.r = r

    def print_m(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.r}/5")


name = input("Movie name: ")
r = input("Rating out of 5: ")

a = Movie(name, r)
a.print_m()
```
