```python
def analyze_post_engagement(likes_list):
    total_likes = 0

    for likes in likes_list:
        total_likes += likes

    print("Total Likes:", total_likes)

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example usage
likes = [200, 150, 300, 250, 150]
analyze_post_engagement(likes)
```

```python
def analyze_rainfall(rainfall_data, required_level):
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    print("Average Rainfall:", int(average))

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example usage
rainfall = [70, 75, 80, 65, 70]
required = 70

analyze_rainfall(rainfall, required)
```

```python
def check_duplicate_usernames(usernames):
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
        print("Status: Flag accounts")
    else:
        print("Duplicate Accounts Found: No")
        print("Status: Accounts are unique")


# Example usage
users = ["rahul", "anita", "rahul", "vikram"]
check_duplicate_usernames(users)
```

```python
def filter_premium_crops(prices):
    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


# Example usage
crop_prices = [1500, 2500, 1800, 3200, 1900]
filter_premium_crops(crop_prices)
```

```python
def check_application_health(errors):
    print("Error Count:", errors)

    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example usage
check_application_health(7)
```

```python
def check_exam_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == 1:
            attended += 1

    percentage = (attended / total_classes) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example usage
attendance = [1, 1, 1, 0, 1]  # 1 = Present, 0 = Absent
check_exam_eligibility(attendance)
```

```python
def calculate_electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example usage
units_consumed = 250
total_bill, usage_status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: â‚¹", total_bill)
print("Usage Classification:", usage_status)
```
