```python
m = [45, 78, 90, 33, 60]

pc = 0
fc = 0

for m in m:
    if m >= 50:
        pc = pc + 1
    else:
        fc = fc + 1

print("Total pass students:", pc)
print("Total fail students:", fc)
```

```python
names = [" Alice ", "bob", " CHARLIE "]

for i in range(len(names)):
  names[i] = names[i].strip().lower()

print("Cleaned names:", names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

count = [0,0,0]

for i in range(len(messages)):
  print(len(messages[i]), messages[i])

for i in range(len(messages)):
  count[i] = len(messages[i])

for i in range(len(count)):
  if(count[i]>10):
    print("Greater than 10 characters:", count[i])
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

count = 0

for i in range(len(logs)):
  if (logs[i] == "ERROR"):
    count += 1

print("Total errors:", count)
```
