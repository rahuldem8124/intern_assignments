```python
marks = []
n = int(input("Enter number of subjects: "))

for i in range(n):
    marks.append(int(input("Enter marks: ")))

avg = 0

for i in range(n):
    avg += marks[i]

avg /= n

if(avg>=50):
  print("Pass avg =",avg)
else:
  print("Fail avg =",avg)
```

```python
def calc_fare(km, peak):
  b_f = 50
  p_km = 12

  fare = b_f + (km*p_km)

  if peak.lower == "true":
    fare = fare*.25

  return fare

while True:
  km = float(input("Enter distance in km:"))
  peak = input("Is it a peak hour? (yes/no):")

  t_f = calc_fare(km, peak)
  print("Total fare:", int(t_f))

  retry = input("Do you want to calculate again? (yes/no): ")
  if retry.lower() != "yes":
    print("Thank you for using cab service ðŸš–")
    break
```

```python
def check_att(at):
  present = at.count("P")
  total = len(at)

  per = (present/total)*100

  if per >= 75:
    print("Eligible")

  else:
    print("Not Eligible")


attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
check_att(attendance)
```

```python
def check_pass(p):
  has_d = False
  has_s = False

  for ch in p:
    if ch.isdigit():
      has_d = True
    if ch in "@#$":
      has_s = True

  if len(p) >=8 and has_s and has_d:
    return"Strong password"

  else:
    return "Weak password"


password = input("Enter password: ")
print(check_pass(password))
```
